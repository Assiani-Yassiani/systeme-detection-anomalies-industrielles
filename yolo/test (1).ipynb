```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import torch


# Vérifier si CUDA est disponible
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
print(f"Device used: {device}")

# Charger le modèle YOLOv8s pré-entraîné
model = YOLO('yolov8s.pt')

# Entraîner le modèle
results = model.train(
    data='data.yaml',    # Chemin vers le fichier de configuration des données
    epochs=120,           # Nombre d'époques d'entraînement
    imgsz=640,           # Taille de l'image pour l'entraînement
    batch=16,
    device=device,          # Taille du lot
    name='yolov8s_run',  # Nom de l'expérience
    save_period=1        # Sauvegarder les poids après chaque époque
)





```
